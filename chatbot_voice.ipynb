```python
# Importations nécessaires
import gradio as gr
import whisper
from transformers import AutoTokenizer, AutoModelForCausalLM
from TTS.api import TTS
import torch
import tempfile
```

```python
# Dans une cellule de code Colab :
!pip install -q huggingface_hub

from huggingface_hub import notebook_login
notebook_login()  # Lance l'interface graphique pour coller le token
```

```python
# Chargement des modèles (à exécuter une fois)
# -------------------------------------------------
# Modèle de transcription
asr_model = whisper.load_model("base")

# Modèle de génération de texte
model_name = "distilgpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
language_model = AutoModelForCausalLM.from_pretrained(model_name)

```

```python
# ✅ STEP 5: Text-to-speech using Coqui TTS
from TTS.api import TTS
import scipy.io.wavfile
import os

# Load Coqui TTS model (CPU-compatible)
tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DDC", progress_bar=False, gpu=False)
```

```python
# Fonction de traitement principale
# -------------------------------------------------
def process_audio(audio_path):
    # Transcription audio -> texte
    asr_result = asr_model.transcribe(audio_path)
    input_text = asr_result["text"]

    # Génération de réponse
    inputs = tokenizer(input_text, return_tensors="pt", max_length=512, truncation=True)
    outputs = language_model.generate(
        inputs.input_ids,
        max_new_tokens=150,
        do_sample=True,
        top_k=50,
        top_p=0.95,
        temperature=0.7
    )
    response_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Synthèse vocale
    with tempfile.NamedTemporaryFile(suffix=".wav", delete=False) as fp:
        tts.tts_to_file(text=response_text, file_path=fp.name)
        return fp.name, input_text, response_text
```

```python
import gradio as gr

# Interface Gradio
# -------------------------------------------------
with gr.Blocks(title="Chatbot Vocal AI") as demo:
    gr.Markdown("# 🎤 Chatbot Vocal Intelligent")
    gr.Markdown("Parlez en français et recevez une réponse vocale!")

    with gr.Row():
        # Utilisation correcte de gr.Audio pour capturer l'audio depuis le microphone
        audio_input = gr.Audio(type="numpy", label="Parlez ici")  # Pas besoin de source="microphone"
        audio_output = gr.Audio(label="Réponse vocale", autoplay=True)

    with gr.Accordion("Détails de la conversation", open=False):
        input_text = gr.Textbox(label="Transcription")
        response_text = gr.Textbox(label="Réponse générée")

    btn = gr.Button("Envoyer")

    # Fonction de traitement de l'audio (à définir)
    def process_audio(audio_input):
        # Logique de traitement de l'audio et génération de la transcription et réponse
        transcription = "Transcription du discours"  # Simule la transcription
        response = "Réponse générée"  # Simule la réponse du chatbot
        return audio_input, transcription, response

    # Déclenchement du processus lors du clic sur le bouton
    btn.click(
        fn=process_audio,
        inputs=audio_input,
        outputs=[audio_output, input_text, response_text]
    )

demo.launch(debug=True, share=True)
```
