```python
# Importations n√©cessaires
import gradio as gr
import whisper
from transformers import AutoTokenizer, AutoModelForCausalLM
from TTS.api import TTS
import torch
import tempfile
```

```python
# Dans une cellule de code Colab :
!pip install -q huggingface_hub

from huggingface_hub import notebook_login
notebook_login()  # Lance l'interface graphique pour coller le token
```

```python
# Chargement des mod√®les (√† ex√©cuter une fois)
# -------------------------------------------------
# Mod√®le de transcription
asr_model = whisper.load_model("base")

# Mod√®le de g√©n√©ration de texte
model_name = "distilgpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
language_model = AutoModelForCausalLM.from_pretrained(model_name)

```

```python
# ‚úÖ STEP 5: Text-to-speech using Coqui TTS
from TTS.api import TTS
import scipy.io.wavfile
import os

# Load Coqui TTS model (CPU-compatible)
tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DDC", progress_bar=False, gpu=False)
```

```python
# Fonction de traitement principale
# -------------------------------------------------
def process_audio(audio_path):
    # Transcription audio -> texte
    asr_result = asr_model.transcribe(audio_path)
    input_text = asr_result["text"]

    # G√©n√©ration de r√©ponse
    inputs = tokenizer(input_text, return_tensors="pt", max_length=512, truncation=True)
    outputs = language_model.generate(
        inputs.input_ids,
        max_new_tokens=150,
        do_sample=True,
        top_k=50,
        top_p=0.95,
        temperature=0.7
    )
    response_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Synth√®se vocale
    with tempfile.NamedTemporaryFile(suffix=".wav", delete=False) as fp:
        tts.tts_to_file(text=response_text, file_path=fp.name)
        return fp.name, input_text, response_text
```
