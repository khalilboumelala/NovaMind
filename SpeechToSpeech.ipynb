```python
# ✅ STEP 2: Upload your audio file
from google.colab import files
uploaded = files.upload()  # Upload a .wav or .mp3 file
filename = list(uploaded.keys())[0]
```

```python
# ✅ STEP 3: Transcribe audio with Whisper
import whisper

asr_model = whisper.load_model("base")  # Or "small", "medium"
asr_result = asr_model.transcribe(filename)
user_input = asr_result["text"]
print("\n🎧 You said:", user_input)
```

```python
# Dans une cellule de code Colab :
!pip install -q huggingface_hub

from huggingface_hub import notebook_login
notebook_login()  # Lance l'interface graphique pour coller le token
```

```python
# ✅ STEP 4: Generate response with Mistral LLM (CPU-compatible)
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name = "mistralai/Mistral-7B-Instruct-v0.1"  # Replace if needed
tokenizer = AutoTokenizer.from_pretrained(model_name)
language_model = AutoModelForCausalLM.from_pretrained(model_name)

prompt = f"[INST] {user_input} [/INST]"
inputs = tokenizer(prompt, return_tensors="pt")

# Force CPU mode
language_model = language_model.to("cpu")
inputs = {k: v.to("cpu") for k, v in inputs.items()}

output_ids = language_model.generate(**inputs, max_new_tokens=100)
response_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)
print("\n🤖 AI Response:", response_text)
```

```python
# ✅ STEP 4: Generate response with a small LLM (distilgpt2)
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name = "distilgpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
language_model = AutoModelForCausalLM.from_pretrained(model_name)

prompt = user_input  # Pas besoin de format spécial

inputs = tokenizer(prompt, return_tensors="pt")
outputs = language_model.generate(**inputs, max_new_tokens=100)
response_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

print("\n🤖 AI Response:", response_text)
```

```python
# ✅ STEP 5: Text-to-speech using Coqui TTS
from TTS.api import TTS
import scipy.io.wavfile
import os

# Load Coqui TTS model (CPU-compatible)
tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DDC", progress_bar=False, gpu=False)

tts_output_path = "response.wav"
tts.tts_to_file(text=response_text, file_path=tts_output_path)
```
